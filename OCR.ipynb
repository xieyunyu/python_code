```python

!pip install  pytesseract
!pip install Pillow
!sudo apt install tesseract-ocr
!sudo apt install libtesseract-dev
!sudo apt install tesseract-ocr-chi-tra

from PIL import Image
import pytesseract

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')


#讀取雲端照片
img_path = "/content/drive/MyDrive/02 CODE/Photo/account1130910(2).jpg"
img = Image.open(img_path)
text = pytesseract.image_to_string(img, lang='chi_tra')
print(text)

```

```python
!pip install aspose-ocr-python-net
import aspose.ocr as ocr


# 將影像新增至辨識批次中
inputImg = ocr.OcrInput(ocr.InputType.SINGLE_IMAGE)
inputImg.add(img)

# 實例化 Aspose.OCR API
api = ocr.AsposeOcr()

# 辨識影像
result = api.recognize(inputImg)

# 列印識別結果
print(result[0].recognition_text)
result[0].save("ImagetoDOCX.docx", ocr.SaveFormat.DOCX);
```
